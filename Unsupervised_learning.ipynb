```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
ds = pd.read_csv("mall-customers.csv")
ds
```

```python
ds.head()
```

KMeans Clustering

```python
x = ds.iloc[:,[3,4]].values  #input variable or independent variable
x
```

```python
#defining kmeans
```

```python
#KMeans===class
```

```python
k = KMeans(n_clusters = 5, init = "k-means++", random_state = 42)#creating an object
k
```

```python
y_pred = k.fit_predict(x)  #dependent or output variable
y_pred   
```

```python
#visualisation
```

```python
plt.scatter(x[y_pred ==0,0], x[y_pred ==0,1], s = 100, c = "blue", label = "Cluster 1 ")
plt.scatter(x[y_pred ==1,0], x[y_pred ==1,1], s = 100, c = "green", label = "Cluster 2 ")
plt.scatter(x[y_pred ==2,0], x[y_pred ==2,1], s = 100, c = "red", label = "Cluster 3 ")
plt.scatter(x[y_pred ==3,0], x[y_pred ==3,1], s = 100, c = "cyan", label = "Cluster 4 ")
plt.scatter(x[y_pred ==4,0], x[y_pred ==4,1], s = 100, c = "magenta", label = "Cluster 5 ")
plt.scatter(k.cluster_centers_[:,0], k.cluster_centers_[:,1],s = 300, c="yellow", label="Centriod")

plt.title("Cluster of customers")
plt.xlabel("Annual income(k$)")
plt.ylabel("Spending score (1-100)")
plt.legend()
plt.show()
```

```python
c = ["blue","green","red","cyan","magenta"]
l = ["Cluster 1","Cluster 2","Cluster 3","Cluster 4","Cluster 5",]
for i in range(5):
    plt.scatter(x[y_pred ==i,0], x[y_pred ==i,1], s = 100, c = c[i], label = l[i])
plt.scatter(k.cluster_centers_[:,0], k.cluster_centers_[:,1],s = 300, c="yellow", label="Centriod")

plt.title("Cluster of customers")
plt.xlabel("Annual income(k$)")
plt.ylabel("Spending score (1-100)")
plt.legend()
plt.show()
```

Hierarchial clustering

```python
#drawing dendrogram
import scipy.cluster.hierarchy as shc
dendro = shc.dendrogram(shc.linkage(x, method = "ward"))
plt.title("Dendrogram plot")
plt.ylabel("Euclidean distances")
plt.xlabel("Customers")
plt.show()
```

Agglomerative clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters = 5, affinity = "euclidean", linkage = "ward")
y1_pred = hc.fit_predict(x)
y1_pred
```

```python
#visualisation
```

```python
c = ["blue","green","red","cyan","magenta"]
l = ["Cluster 1","Cluster 2","Cluster 3","Cluster 4","Cluster 5",]
for i in range(5):
    plt.scatter(x[y1_pred ==i,0], x[y1_pred ==i,1], s = 100, c = c[i], label = l[i])

plt.title("Cluster of customers")
plt.xlabel("Annual income(k$)")
plt.ylabel("Spending score (1-100)")
plt.legend()
plt.show()
```
